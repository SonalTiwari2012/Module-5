```python
#Sorting Data using pandas
data= {'Name':['Shreya','Pooja','Anushka','Nitya','Shreya','Sonia'],
      'Age':[12,13,14,15,16,17],
       'Salary':[70000,80000,50000,60000,40000,20000]}
df=pd.DataFrame(data)
print(df)
```

```python
# Sorting data by index
df=pd.DataFrame(data,index=['a','b','c','d','e','f'])
print(df)
```

```python
#Sorting data by column
xyz=df.sort_index(axis=1)
print(xyz)
```

```python
st=df.sort_values(by='Age')
print(st)
```

```python
#Grouping in pandas
data={
    'Department':['Finance','HR','HR','Trainer','IT','IT','Sales','Finance'],
    'Employee':['Atiya','Anushka','Antima','Shivani','Sonia','Pooja','Shubhangi','Shreya'],
    'Salary':[1000,2000,1500,4000,3500,4200,3700,5000],
    'Age':[20,21,22,23,24,25,26,24]}
 
df=pd.DataFrame(data)
print(df)
    
```

```python
x=df.groupby(['Department']).mean(numeric_only=True)
print(x)
```

```python
y=df.groupby('Department').agg({
     'Salary':['sum','mean'],
})
print(y)
```

```python
y=df.groupby('Department').agg({
     'Salary':['sum','mean'],
    'Age':['min','max'],
})
print(y)
```
